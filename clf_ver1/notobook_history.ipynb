```python
import pandas as pd

def show_history(filename, mode, sort_by=['Accuracy']):
    def _color_red_or_green(value):
        if isinstance(value, bool) and value == True:
            color = 'red'
        elif isinstance(value, bool) and value == False:
            color = 'black'
        elif isinstance(value, str):
            color = 'green'
        else:
            color = 'gray'
        return 'color: %s' % color

    history = pd.read_csv(filename)
    history = history[history["Mode"] == mode]
    history = history.sort_values(by=sort_by, ascending=False).T
    if mode == "Train":
        history.columns += 1
    history = history.style.applymap(_color_red_or_green)
    return history
```

```python
show_history("history.csv", mode="Train", sort_by=['Prec.', 'AUC'])
```

```python
show_history("history.csv", mode="Test", sort_by=['Prec.', 'AUC'])
```
